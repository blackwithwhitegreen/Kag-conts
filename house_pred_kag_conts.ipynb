```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/house/train.csv')
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
df_train.shape
```

```python
y = df_train['SalePrice']
fea = ['LotArea','MiscVal','MoSold','YrSold','PoolArea','EnclosedPorch','OpenPorchSF','WoodDeckSF','KitchenAbvGr','FullBath',
             'TotalBsmtSF','YearBuilt',]
X = df_train[fea]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
rf = DecisionTreeRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
pred  = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mea = mean_absolute_error(y_test,pred)
print(f"Mean abs error : {mea}")
```

```python
df_test = pd.read_csv('/content/drive/MyDrive/house/test.csv')
```

```python
df_test.head()
```

```python
X_test = df_test[fea]
```

```python
test_pred = rf.predict(X_test)
```

```python
res = pd.DataFrame({'Id' : df_test.Id,'SalePrice': test_pred})
res.to_csv('submission.csv',index = False)
print('Submission file saved as submission.csv')
```
