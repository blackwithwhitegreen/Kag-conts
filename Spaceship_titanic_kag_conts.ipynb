#This is notebook for the kaggle-competition(spaceship-titanic).
#Using of the DecisionTree classifier,and getting the prediction.
#At last saving the submission in the Submission.csv file.

```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('/content/drive/MyDrive/spaceship-titanic/train.csv')
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
y = df_train['Transported']
fea = ['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
X = df_train[fea]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
pred = dt.predict(X_test)
```

```python
dt.score(X_test,y_test)
```

```python
df_test = pd.read_csv('/content/drive/MyDrive/spaceship-titanic/test.csv')
```

```python
df_test.head()
```

```python
X_test = df_test[fea]
```

```python
test_pred = dt.predict(X_test)
```

```python
res = pd.DataFrame({'PassengerId' : df_test.PassengerId 	,'Transported': test_pred})
res.to_csv('submission.csv',index = False)
print('Submission file saved as submission.csv')
```
